```python
from datasets import Dataset
```

```python
ds = Dataset.from_generator(DataSetRowGen)
```

```python
ds
```

```python
ds = ds.remove_columns('id')
ds = ds.rename_column("desc", "text")
ds
```

```python
tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True)
```

```python
ds = ds.map(
    preprocess_function,
    batched=True,
    num_proc=4,
)
```

```python
ds_train_test = ds.train_test_split(train_size=0.85, seed=13)
ds_train_valid_test = ds_train_test['train'].train_test_split(train_size=0.83, seed=31)
ds_train_valid_test['validation'] = ds_train_valid_test.pop("test")
ds_train_valid_test['test'] = ds_train_test['test']
ds_train_valid_test
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False, return_tensors="tf")
```

```python
from transformers import create_optimizer, AdamWeightDecay

optimizer = AdamWeightDecay(learning_rate=2e-5, weight_decay_rate=0.01)
```

```python
from transformers import TFAutoModelForCausalLM

model = TFAutoModelForCausalLM.from_pretrained("distilgpt2")
```

```python
tf_train_set = model.prepare_tf_dataset(
    ds_train_valid_test["train"],
    shuffle=True,
    batch_size=16,
    collate_fn=data_collator,
)

tf_test_set = model.prepare_tf_dataset(
    ds_train_valid_test["test"],
    shuffle=False,
    batch_size=16,
    collate_fn=data_collator,
)
```

```python
model.fit(x=tf_train_set, validation_data=tf_test_set, epochs=3)
```
